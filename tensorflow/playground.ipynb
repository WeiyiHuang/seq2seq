```python
def batch_iter(data, batch_size, num_epochs, shuffle=True):
    """
    Generates a batch iterator for a dataset.
    """
    data = np.asarray(data)
    print(data)
    print(data.shape)
    data_size = len(data)
    num_batches_per_epoch = int(len(data)/batch_size) + 1 
    for epoch in range(num_epochs):
        # Shuffle the data at each epoch
        if shuffle:
            shuffle_indices = np.random.permutation(np.arange(data_size))
            shuffled_data = data[shuffle_indices]
        else:
            shuffled_data = data
        for batch_num in range(num_batches_per_epoch):
            start_index = batch_num * batch_size
            end_index = min((batch_num + 1) * batch_size, data_size)
            yield shuffled_data[start_index:end_index]
```

```python
batches = batch_iter([[0,1,1], [3,2,1], [3,2,4], [4,3,2], [6,1,2]], 2, 3)
```

```python
h1 = tf.placeholder(tf.float32, shape=(None, 3, 4))
h2 = tf.placeholder(tf.float32, shape=(None, 3, 4))
h_q = tf.placeholder(tf.float32, shape=(None, 3, 4))
h_a = tf.placeholder(tf.float32, shape=(None, 3, 4))
h_q_mix = tf.concat([h1, h_q], -1)
h_a_mix = tf.concat([h2, h_a], -1)
reshape_a_t = tf.transpose(h_a_mix, [0, 2, 1])
S_qa_raw = tf.matmul(h_q_mix, reshape_a_t)
# reshape_S_qa = tf.reshape(S_qa_raw, [-1, 9])
# reshape_S_qa = tf.nn.softmax(reshape_S_qa)
# S_qa = tf.reshape(reshape_S_qa, [-1, 3, 3])
S_qa = tf.nn.softmax(S_qa_raw, dim=2)
S_aq = tf.transpose(S_qa, [0, 2, 1])
h_a_sum = tf.matmul(S_aq, h_q_mix)
h_q_sum = tf.matmul(S_qa, h_a_mix)
h_q_coa = tf.matmul(S_qa, h_a_sum)
h_a_coa = tf.matmul(S_aq, h_q_sum)
h_q_final = tf.concat([h_q_mix, h_a_sum, h_q_coa], -1)[:, -1, :]
h_a_final = tf.concat([h_a_mix, h_q_sum, h_a_coa], -1)[:, -1, :]
```

```python
with tf.Session() as sess:
    np.random.seed(10)
    rand_array = np.random.rand(2, 3, 4)
    print(sess.run(h_q_final, feed_dict={h1: rand_array, h2: rand_array, h_q: rand_array, h_a: rand_array}))  # Will succeed.
```
